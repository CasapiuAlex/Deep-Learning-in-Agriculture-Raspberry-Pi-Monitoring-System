# Clasificare

```python
import cv2
import numpy as np
import tensorflow as tf

# Căile către fișiere
VIDEO_PATH = '/content/Test_Final_Clasificare.mp4'
MODEL_PATH = '/content/best_leaf_model (8).keras'
LABELS_PATH = '/content/labels.txt'  # Un fișier text cu numele claselor, fiecare pe o linie nouă
OUTPUT_VIDEO_PATH = 'video_cu_clasificari.mp4'

# Parametrii modelului și ai procesării
MODEL_INPUT_SIZE = (256, 256) # Dimensiunea de intrare
CLASSIFY_INTERVAL_SEC = 4.0   # Intervalul la care se face clasificarea
PAUSE_DURATION_SEC = 2.0      # Durata pentru care cadrul "îngheață" pentru a afișa rezultatul

print("Se încarcă modelul Keras...")
model = tf.keras.models.load_model(MODEL_PATH)
print("Model încărcat.")

print("Se încarcă etichetele (clasele)...")
with open(LABELS_PATH, 'r') as f:
    class_labels = [line.strip() for line in f.readlines()]
print(f"Clase găsite: {class_labels}")

# Deschidem fișierul video de intrare
cap = cv2.VideoCapture(VIDEO_PATH)
if not cap.isOpened():
    print("Eroare: Nu s-a putut deschide fișierul video.")
    exit()

# Obținem proprietățile videoclipului (FPS, dimensiuni)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Definim codecul și creăm obiectul VideoWriter pentru a salva videoclipul de ieșire
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (frame_width, frame_height))

print(f"\nSe procesează videoclipul: {VIDEO_PATH}")
print(f"FPS: {fps}, Dimensiuni: {frame_width}x{frame_height}")

frame_idx = 0
last_classify_time = -CLASSIFY_INTERVAL_SEC # Forțăm clasificarea primului cadru

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break # Am ajuns la finalul videoclipului

    current_time_sec = frame_idx / fps

    # Verificăm dacă a trecut suficient timp pentru o nouă clasificare
    if current_time_sec >= last_classify_time + CLASSIFY_INTERVAL_SEC:
        last_classify_time = current_time_sec

        # 1. Preprocesarea cadrului pentru model
        img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img_resized = cv2.resize(img_rgb, MODEL_INPUT_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img_resized)
        img_array = np.expand_dims(img_array, axis=0) # Creăm un batch de 1
        img_preprocessed = img_array / 255.0 # Normalizare

        # 2. Realizarea predicției
        predictions = model.predict(img_preprocessed)
        score = np.max(predictions[0])
        class_id = np.argmax(predictions[0])
        label = class_labels[class_id]

        # 3. Afișarea rezultatului pe cadru
        text = f"Clasa: {label} ({score:.2f})"
        print(f"Timp: {current_time_sec:.2f}s - {text}")
        # Punem textul pe o copie a cadrului
        frame_with_text = frame.copy()
        cv2.putText(frame_with_text, text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (105, 105, 105), 3)

        # 4. "Înghețarea" cadrului - scriem același cadru de mai multe ori
        pause_frames = int(PAUSE_DURATION_SEC * fps)
        for _ in range(pause_frames):
            out.write(frame_with_text)
    else:
        # Dacă nu este timpul pentru clasificare, scriem cadrul original
        out.write(frame)

    frame_idx += 1
    # Afișăm progresul
    if frame_idx % 100 == 0:
        print(f"Procesat cadru {frame_idx}/{total_frames}...")

# --- PASUL 5: FINALIZARE ---
print("\nProcesare finalizată.")
cap.release()
out.release()
cv2.destroyAllWindows()
print(f"Videoclipul rezultat a fost salvat la: {OUTPUT_VIDEO_PATH}")
```
